```python
import kagglehub
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


sns.set()
```

```python
path = kagglehub.dataset_download("gregorut/videogamesales")
```

```python
df = pd.read_csv(path + "/vgsales.csv")
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Year'], bins=20)
plt.title("Distribution Video Game Release years")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Global_Sales'], bins=20)
plt.title("Total worldwide sales")
plt.show()
```

```python
country_revenue = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].sum()

plt.figure(figsize=(10,5))
country_revenue.plot(kind='bar')
plt.title("Total Revenue by Region")
plt.ylabel("Total Sales (Millions)")
plt.show()

country_revenue
```

```python
df.columns
```

```python
genre_revenue = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,5))
genre_revenue.plot(kind='bar')
plt.title("Total Global Sales by Genre")
plt.ylabel("Total Sales (Millions)")
plt.show()

genre_revenue
```

```python
genre_revenue = df.groupby('Genre')['Global_Sales'].sum().sort_values(ascending=False)

genre_revenue
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='NA_Sales', y='EU_Sales', data=df)
plt.title("NA Sales vs EU Sales")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.heatmap(df[['Rank', 'Global_Sales', 'NA_Sales', 'EU_Sales']].corr(), annot=True)
plt.title("Correlation Matrix")
plt.show()
```
